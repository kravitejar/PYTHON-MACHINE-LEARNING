```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as nm
```

```python
dataset=pd.read_csv(r'C:\Users\RAVI\Desktop\diabetes.csv')
dataset
```

```python
zero_not_accepted = ['Glucose','BloodPressure','SkinThickness','BMI','Insulin']
```

```python
for col in zero_not_accepted:
    dataset[col]=dataset[col].replace(0,nm.NaN)
    mean = int(dataset[col].mean(skipna=True))
    dataset[col] = dataset[col].replace(nm.NaN, mean)
```

```python
X = dataset.iloc[:,0:8]
y = dataset.iloc[:,8]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
#Calculate 'k'
import math

len(X_test)
```

```python
print(math.sqrt(len(X_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')

classifier.fit(X_train,y_train)

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

cm = confusion_matrix(y_test,y_pred,labels=[0,1])
print( cm)
```

```python
print (pd.crosstab(y_test,y_pred,rownames=['Actual Values'],colnames=['Predicted Values']))
```

```python
print (f1_score(y_test,y_pred))
print ('Accuracy Percentage: ',accuracy_score(y_test,y_pred))
```
